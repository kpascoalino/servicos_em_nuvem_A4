```python
agents = pd.read_csv('/content/AGENTS_KELLY.csv')
customer = pd.read_csv('/content/CUSTOMER_KELLY.csv')
orders = pd.read_csv('/content/ORDERS_KELLY.csv')
```

```python
# vamos fazer uma seleção do código dos agentes cujo valor é instanciado mais de uma vez na tabela ORDERS, apresentando a contagem
aux = pd.DataFrame(orders['AGENT_CODE'].value_counts()).reset_index()
select_count = aux[aux['count'] > 1].sort_values(by=['count'], ascending=False)
display(select_count)
```

```python
# agora que sabemos os agentes que mais venderam, queremos saber se os clientes para os quais realizaram as vendas (A002 e A010) foram sempre os mesmos
select_join = agents.merge(orders, how='outer', on='AGENT_CODE')
select_join = select_join.merge(customer, how='outer', on='CUST_CODE')
select_join = select_join[['CUST_CODE', 'AGENT_CODE_x']]
select_join = select_join = select_join.query("AGENT_CODE_x == 'A002' or AGENT_CODE_x == 'A010'")
select_join = select_join.rename(columns={'AGENT_CODE_x': 'AGENT_CODE'})
select_join['ORDER_AMT'] = 1
select_join = select_join.groupby(['AGENT_CODE', 'CUST_CODE']).sum()
select_join
```
